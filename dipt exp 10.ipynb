```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = np.zeros((500, 500, 3), dtype=np.uint8)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image1, 'NIKSHITHA', (100, 250), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
```

```python
kernel = np.ones((3, 3), np.uint8)
```

```python
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for displaying
plt.title("Input Image with Text")
plt.axis('off')
```

```python
opened_image = cv2.morphologyEx(image1, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(cv2.cvtColor(opened_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Opening Operation")
plt.axis('off')
```

```python
closed_image = cv2.morphologyEx(image1, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(cv2.cvtColor(closed_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Closing Operation")
plt.axis('off')
```
